```python
import pandas as pd
import numpy as np
train = pd.read_csv('./data/train.csv')
```

é¦–å…ˆï¼Œç›®æ ‡å°±æ˜¯å½¢æˆç±»ä¼¼äºä¸‹é¢çš„è¿™æ ·çš„è¡¨ï¼ˆ<font color =#ff7f01 >èƒ½åšæˆçº¢æ¡†é‡Œè¿™æ ·å°±æ˜¯æ»¡åˆ†5åˆ†ï¼Œè¦æ˜¯èƒ½æŠŠå³è¾¹ä¸¤åˆ—ä¹Ÿåšå‡ºæ¥ï¼Œæˆ‘ä¼šç»™8åˆ†</font>ï¼‰

ã€è¿™ä¸ªè¡¨æœ‰59è¡Œï¼Œç›¸å½“äºå¯¹æ•°æ®çš„æ¯ä¸€ä¸ªç»´åº¦è¿›è¡Œæ¢³ç†åˆ†æã€‘

é‚£ä¹ˆï¼Œå˜é‡ç±»å‹çš„é€»è¾‘ï¼Œå¸®å¤§å®¶æ¢³ç†ä¸€ä¸‹ï¼š 
1. åˆ—åé‡Œæœ‰`bin`çš„æˆ–è€…è¿™ä¸€åˆ—æ˜¯targetï¼Œæˆ‘ä»¬å½’ç±»ä¸º`binary` (if)
2. åˆ—åé‡Œæœ‰`cat`çš„æˆ–è€…è¿™ä¸€åˆ—æ˜¯idï¼Œå½’ç±»ä¸º`nominal` (elif)
3. æ•°æ®ç±»å‹ä¸º`float`çš„æ˜¯`interval` (elif)
4. æ•°æ®ç±»å‹æ˜¯`int`çš„æ˜¯`ordinal` (elif)

> ordinal è§£é‡Šä¸€ä¸‹ï¼šåˆ¨é™¤å¼€å‰é¢çš„æƒ…å†µï¼Œå‰©ä¸‹çš„ç‰¹å¾ï¼Œå¦‚æœä»–ä»¬çš„æ•°æ®ç±»å‹æ˜¯`int`ï¼Œæˆ‘ä»¬è®¤ä¸ºä»–ä»¬æ˜¯ordinalçš„ï¼Œå› ä¸ºæ•´æ•°çš„æ’åˆ—æ˜¯æœ‰åºçš„ï¼Œæˆ‘ä»¬è®¤ä¸º2>1ï¼Œ3>2ç­‰ç­‰ï¼Œå¯¹åº”å€¼çš„å½±å“ç¨‹åº¦ä¹Ÿæ»¡è¶³è¿™æ ·çš„åºåˆ—å…³ç³»ï¼ˆè™½ç„¶æˆ‘ä»¬ä¸çŸ¥é“å€¼çš„å…·ä½“å«ä¹‰ï¼‰

```python
# å¤§å®¶å°è¯•ä¸€ä¸‹å§~

# ä»£ç å†™åœ¨ä¸‹æ–¹ï¼Œç›®æ ‡ç”Ÿæˆä¸€ä¸ªdfï¼Œå‘½åä¸ºmeta


data = []
for col in train.columns:
    # Defining the role
    if col == 'target':
        role = 'ç›®æ ‡æ ‡ç­¾ï¼ˆæ˜¯å¦å‘èµ·ç´¢èµ”ï¼‰'
    elif col == 'id':
        role = 'åºåˆ—å·ï¼ˆä¸å±äºç‰¹å¾ï¼‰'
    else:
        role = 'ç‰¹å¾'
         
    # Defining the level
    if 'bin' in col or col == 'target':
        level = 'binary'
    elif 'cat' in col or col == 'id':
        level = 'nominal'
    elif train[col].dtype == 'float64':
        level = 'interval'
    elif train[col].dtype == 'int64':
        level = 'ordinal'
        
    # Initialize keep to True for all variables except for id
    keep = True
    if col == 'id':
        keep = False
        
    # Defining the data type 
    dtype = train[col].dtype
    
    # Creating a Dict that contains all the metadata for the variable
    col_dict = {
        'å˜é‡åç§°': col,
        'æ•°æ®è§’è‰²': role,
        'å˜é‡ç±»å‹': level,
        'æ•°æ®ç±»å‹': dtype,
        'æ˜¯å¦ä¿ç•™': keep,
    }
    data.append(col_dict)
    
meta = pd.DataFrame(data, columns=['å˜é‡åç§°', 'æ•°æ®è§’è‰²', 'å˜é‡ç±»å‹', 'æ•°æ®ç±»å‹', 'æ˜¯å¦ä¿ç•™'])
meta.set_index('å˜é‡åç§°', inplace=True)
meta['å”¯ä¸€å€¼æ•°é‡'] = [train.replace(-1,np.nan)[col].value_counts().shape[0] if col in meta[(meta['å˜é‡ç±»å‹'] != 'interval') & (meta['æ˜¯å¦ä¿ç•™'])].index else 'ä¸ç»Ÿè®¡' for col in meta.index]
meta['ç¼ºå¤±å€¼æ•°é‡'] = [str(train[train[col] == -1][col].count()) + f'({round(train[train[col] == -1][col].count()/train.shape[0]*100,2)}%)' for col in meta.index]
meta
```

where ~r = desired~proportion~rate

<div class="alert alert-block alert-success"><b>Step 1 (5åˆ†)</b>: 
    
æ ¹æ®æŒ‡ç¤ºå®Œæˆä¸‹é¢çš„å‡½æ•°`undersampling`ã€‚

```python
def undersampling(df, desired_prop_rate = 0.15, plot = True):
    # è·å–target=0å’Œ1çš„index
    idx_class_0 = df[df.target == 0].index
    idx_class_1 = df[df.target == 1].index
    
    # è·å–target=0å’Œ1çš„ä¸ªæ•°
    count_class_0 = len(idx_class_0)
    count_class_1 = len(idx_class_1)
    
    # æ ¹æ®ä¸Šé¢çš„å…¬å¼è®¡ç®—undersampling rate
    undersampling_rate = ((1-desired_prop_rate)*count_class_1)/(count_class_0*desired_prop_rate)
    undersampled_majority_size = int(undersampling_rate*count_class_0)
    
    print(f"å…³äºtarget=0çš„æ¬ é‡‡æ ·æ¯”ç‡ä¸º:ã€{undersampling_rate}ã€‘")
    print(f"åœ¨æ¬ é‡‡æ ·ä¹‹åï¼Œtarget=0çš„æ•°é‡ä¸ºï¼šã€{undersampled_majority_size}ã€‘")
    
    
    from sklearn.utils import shuffle
    # ç”¨shuffleå‡½æ•°å¯¹target=0çš„æ‰€æœ‰æ•°æ®æ‰“ä¹±æŠ½æ ·ï¼Œseed=100,æŠ½æ ·æ•°ä¸ºåˆšåˆšè®¡ç®—çš„å€¼
    undersampled_idx = shuffle(idx_class_0, random_state=100, n_samples=undersampled_majority_size)
    
    # æŠŠundersamplingä¹‹åçš„idxå’Œä¹‹å‰çš„target=1çš„idxåˆå¹¶ï¼Œå¹¶åœ¨trainé‡Œé¢æ ¹æ®idxæŠŠæ•°æ®ç´¢å¼•å‡ºæ¥
    idx_total = list(undersampled_idx)+list(idx_class_1)
    
    df = df.loc[idx_total].reset_index(drop=True)
    
    # å¦‚æœplotå‚æ•°è®¾ç½®ä¸ºTrueï¼Œç”»å‡ºæ¬ é‡‡æ ·ä¹‹åçš„targetåˆ†å¸ƒæŸ±å½¢å›¾ã€‚
    if plot == True:
        import seaborn as sns
        sns.countplot(x = df.target)
        # print(df.target.value_counts()/df.shape[0])
    
    return df
```

```python
def add_noise(series, noise_level):
    return series * (1 + noise_level * np.random.randn(len(series)))

def target_encode(trn_series=None, 
                  tst_series=None, 
                  target=None, 
                  min_samples_leaf=1, 
                  smoothing=1,
                  noise_level=0):
    """
    Smoothing is computed like in the following paper by Daniele Micci-Barreca
    https://kaggle2.blob.core.windows.net/forum-message-attachments/225952/7441/high%20cardinality%20categoricals.pdf
    trn_series : training categorical feature as a pd.Series
    tst_series : test categorical feature as a pd.Series
    target : target data as a pd.Series
    min_samples_leaf (int) : minimum samples to take category average into account
    smoothing (int) : smoothing effect to balance categorical average vs prior  
    """ 
    assert len(trn_series) == len(target)
    assert trn_series.name == tst_series.name
    temp = pd.concat([trn_series, target], axis=1)
    # é›†æ•£è¿™ä¸€åˆ—å…³äºtargetçš„group mean
    averages = temp.groupby(by=trn_series.name)[target.name].agg(["mean", "count"])
    # å¹³æ»‘
    smoothing = 1 / (1 + np.exp(-(averages["count"] - min_samples_leaf) / smoothing))
    # è®¡ç®—æ‰€æœ‰targetçš„mean
    prior = target.mean()
    # The bigger the count the less full_avg is taken into account
    # å¦‚æœæŸä¸€ç±»åˆ«çš„å€¼çš„ä¸ªæ•°ç‰¹åˆ«å¤šï¼Œæ¯”å¦‚104è¿™ä¸€ç±»å°±æœ‰21255è¡Œï¼Œé‚£ä¹ˆæˆ‘ä»¬å°±è¦å‰Šå‡å…¶å…³äºtargetçš„meançš„æƒå€¼ã€‚
    averages[target.name] = prior * (1 - smoothing) + averages["mean"] * smoothing
    averages.drop(["mean", "count"], axis=1, inplace=True)
    
    # åˆ†åˆ«å¯¹trainå’Œtestè®¡ç®—mean
    ft_trn_series = pd.merge(
        trn_series.to_frame(trn_series.name),
        averages.reset_index().rename(columns={'index': target.name, target.name: 'average'}),
        on=trn_series.name,
        how='left')['average'].rename(trn_series.name + '_mean').fillna(prior)

    ft_trn_series.index = trn_series.index 
    ft_tst_series = pd.merge(
        tst_series.to_frame(tst_series.name),
        averages.reset_index().rename(columns={'index': target.name, target.name: 'average'}),
        on=tst_series.name,
        how='left')['average'].rename(trn_series.name + '_mean').fillna(prior)

    ft_tst_series.index = tst_series.index
    return add_noise(ft_trn_series, noise_level), add_noise(ft_tst_series, noise_level)
```

```python
def meta(raw_data, processed_data=None, cols_to_drop = [], keep_missing = True):
    """
    raw_data: raw data obtained
    data_processed: data after preprocessing
    cols_to_drop: excluded cols that contain too many missing values or we dont use
    keep_missing: True if we find we dont need to eliminate missing values after imputation
    """
    data = []
    if processed_data is not None:
        df = processed_data.copy()
    else:
        df = raw_data
    for col in df.columns:
        # Defining the role
        if col == 'target':
            role = 'ç›®æ ‡æ ‡ç­¾ï¼ˆæ˜¯å¦å‘èµ·ç´¢èµ”ï¼‰'
        elif col == 'id':
            role = 'åºåˆ—å·ï¼ˆä¸å±äºç‰¹å¾ï¼‰'
        else:
            role = 'ç‰¹å¾'

        # åŠ å…¥ratioå˜é‡
        if 'bin' in col or col == 'target':
            level = 'binary'
        elif 'cat' in col[-3:] or col == 'id':
            level = 'nominal'
        elif df[col].dtype == 'float64' and df[col].replace(-1,np.nan).max()-df[col].replace(-1,np.nan).min() > 1:
            level = 'interval'
        elif df[col].dtype == 'float64' and df[col].replace(-1,np.nan).max()-df[col].replace(-1,np.nan).min() <= 1:
            level = 'ratio'
        elif df[col].dtype == 'int64':
            level = 'ordinal'

        # Initialize keep to True for all variables except for id
        keep = True
        if col == 'id' or col in cols_to_drop:
            keep = False

        # Defining the data type 
        dtype = df[col].dtype
        
        # å®šä¹‰å”¯â¼€å€¼æ•°é‡
# å®šä¹‰å”¯â¼€å€¼æ•°é‡
        if col == 'id' or df[col].dtype == 'float64':
            uniq = 'ä¸ç»Ÿè®¡'
        else:
            if keep_missing == True:
                uniq = train[col].nunique()
            else:
                uniq = train[col].replace({-1:np.nan}).nunique()

        # å®šä¹‰ç¼ºå¤±å€¼
        n_missing = sum(df[col] == -1)
        missing = f'{n_missing}({round(n_missing*100/len(train),2)}%)'
    

        col_dict = {
            'å˜é‡åç§°': col,
            'æ•°æ®è§’è‰²': role,
            'å˜é‡ç±»å‹': level,
            'æ•°æ®ç±»å‹': dtype,
            'æ˜¯å¦ä¿ç•™': keep,
            'å”¯ä¸€å€¼æ•°é‡': uniq,
            'ç¼ºå¤±å€¼æ•°é‡': missing
        }
        data.append(col_dict)
    
    meta = pd.DataFrame(data, columns=list(col_dict.keys()))
    meta.set_index('å˜é‡åç§°', inplace=True)
    
    missing_cols = [col for col in raw_data.columns if raw_data[col].min() == -1]
    meta['ç¼ºå¤±å€¼å¤„ç†'] = ['æ— ç¼ºå¤±å€¼' if col not in missing_cols
        else 'å·²å»é™¤' if col in cols_to_drop
        else 'ä¿ç•™ç¼ºå¤±å€¼' if meta.loc[col,'ç¼ºå¤±å€¼æ•°é‡'][0] != '0' 
        else 'å·²å¡«å……' for col in meta.index]

    
    return meta
```
