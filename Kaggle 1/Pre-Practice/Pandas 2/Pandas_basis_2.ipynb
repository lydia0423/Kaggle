```python
df['Species_encoding'] = df['Species']
df['Species_encoding'].replace(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
                        [1, 2, 3], inplace=True)
df
```

```python
# did not meet the requirements, should use seed = 1
# iris_sample = df.groupby('Species').apply(lambda x: x.sample(5)).reset_index(drop=True)
# iris_sample

setosa_sample = df[df.Species_encoding == 1].sample(n=5, random_state=1)
versicolor_sample = df[df.Species_encoding == 2].sample(n=5, random_state=1)
virginica_sample = df[df.Species_encoding == 3].sample(n=5, random_state=1)

iris_sample = pd.concat([setosa_sample, versicolor_sample, virginica_sample]).reset_index()
iris_sample
```
